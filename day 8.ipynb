```python
# FEATURE SCALING :------>STANDARDIZATION
#                  ------>NORMALIZATION
```

```python
# STANDARDIZATION : it is a very effective technique which re-Scales a feature value so that it has
#                   distribution with 0 mean value and varience equals to 1.

# FEATURE : Xnew = (Xi-Xmean)/Standard Deviation
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Loan_Data.csv")
```

```python
df
```

```python
df.head(3)
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
df["ApplicantIncome"].fillna(df["ApplicantIncome"].mean())
```

```python
df["ApplicantIncome"].fillna(df["ApplicantIncome"].mean(),inplace=True)
```

```python
sns.distplot(df["ApplicantIncome"])
plt.show()
```

```python
df.describe()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
ss.fit(df[["ApplicantIncome"]])
```

```python
ss.transform(df[["ApplicantIncome"]])
```

```python
df["ApplicantIncome_ss"] = pd.DataFrame(ss.transform(df[["ApplicantIncome"]]),columns=["x"])
```

```python
df.head(3)
```

```python
df.describe()
```

```python
plt.figure(figsize=(10,7))
plt.subplot(1,2,1)
plt.title("Before")
sns.distplot(df["ApplicantIncome_ss"])
plt.subplot(1,2,2)
plt.title("After")
sns.distplot(df["ApplicantIncome"])

plt.show()
```

```python
plt.figure(figsize=(10,7))

# Subplot 1
plt.subplot(1, 2, 1)
plt.title("Before")
sns.histplot(df["ApplicantIncome_ss"], kde=True)

# Subplot 2
plt.subplot(1, 2, 2)
plt.title("After")
sns.histplot(df["ApplicantIncome"], kde=True)

plt.tight_layout()
plt.show()
```

```python
df.info()
```

```python
# FEATURE SCALING NORMALIZATION : 
# MIN-MAX SCALER (NORMALIZATION TECHNIQUES) 
```

```python
# NORMALIZATION : it is a  scaling technique in which values are shifted and rescaled so that
#                 they end up ranging between 0 and 1. it is also known as min-max scaling.

# FORMULA : Xnew = Xi-min(X)/max(x)-min(x)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Loan_Data.csv")
```

```python
df
```

```python
df.head(3)
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
sns.distplot(df["CoapplicantIncome"])
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
ms = MinMaxScaler()
ms.fit(df[["CoapplicantIncome"]])
```

```python
ms.transform(df[["CoapplicantIncome"]])
```

```python
df["CoapplicantIncome_min"] = ms.transform(df[["CoapplicantIncome"]])
```

```python
df.head(3)
```

```python
plt.subplot(1,2,1)
plt.title("Befor")
sns.distplot(df["CoapplicantIncome"])

plt.subplot(1,2,2)
plt.title("After")
sns.distplot(df["CoapplicantIncome_min"])

plt.show()
```

```python
plt.figure(figsize=(10,7))

# Subplot 1
plt.subplot(1, 2, 1)
plt.title("Before")
sns.histplot(df["CoapplicantIncome"], kde=True)

# Subplot 2
plt.subplot(1, 2, 2)
plt.title("After")
sns.histplot(df["CoapplicantIncome_min"], kde=True)

plt.tight_layout()
plt.show()
```
