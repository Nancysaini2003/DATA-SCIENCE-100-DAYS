```python
# MACHINE LEARNING : is the science of getting computers to learn and act like humans do, and improve their
#                    learning over time in autonomous fashion , by feeding them data and information in the 
#                    form of observation and real-world intractions.

# HOW DOES MACHINE LEARNING WORK : A machine learning systems learns form historical data,the prediction models,and 
#                                  whenever it receives new data, predicts the output for it.

# input past data ---->training---->machine learning algorithm(LEARN FROM DATA) ----->building local models(NEW DATA)---->output
# DATA(input) -----> TARDITIONAL PROGRAMMING ------> OUTPUT
# PROGRAM --------> TARDITIONAL PROGRAMMING ------> OUTPUT

# DATA(input) ------> MACHINE LEARNING -------> PROGRAM
# OUTPUT --------> MACHINE LEARNING -------> PROGRAM
```

```python
# CLASSIFICATION OF MACHINE LEARNING.
# THERE ARE A THREE TYPE OF MACHINE LEARNING :
# 1) SUPERVISED MACHINE LEARNING
# 2) UNSUPERVISED MACHINE LEARNING
# 3) REINFORCEMENT MACHINE LEARNING

# ADVANTAGES OF MACHINE LEARNING :
# 1) easy identifies trends and patterns
# 2) no human interventaion needed(automation)
# 3) continuous improvement
# 4) handling multi-dimensional and multi-variety data
# 5) wide applications

# DISADVANTAGES OF MACHINE LEARNING :
# 1) data acquisition 
# 2) time and resources
# 3) interpretation of results
# 4) high error-susceptibility

# USES OF MACHINE LEARNING :
# automatic language translation
# medical diagnosis
# stock market trading
# online fraud detection
# virtual personal assiatant
# email span & malware filtering
# self driving cars
# product recommendation
# traffic prediction 
# speech recognition
# image recognition
```

```python
# MACHINE LEARNING ROADMAP :
#                                                                                 1) END TO END ML PROJECTS
#                                                                                 2) MODEL DEPLOYMENTS
#                                             # FEATURE SELECTION                                   3) DOCKERS AND KUBERNETES
#                                             1)  CORRELATION                     4) GRIDSEARCH,RANDOMISEDSEARCH,HYPEROPT,GENTECS ALGO.
#                                             2)  FORWARD ELIMINATION             5) HYPER PARAMETER TUNING
# # FEATURE ENGINEERING                       3)  BACKWARD ELIMINATION            6) LINEAR,LOGISTIC REGRESSION,DECISION TREE,RANDOM FOREST,KMEANS
# 1) EXPLORATARY DATA ANALYSIS                4)  UNIVARIATE SELECTION            7) ML ALGO. , REGRESSION AND CLASSIFICATION,CLUSTERING
# 2) HANDLING MISSING VALUE                   5)  RANDOM FOREST IMPORTANCE        8) FEATURE SELECTION
# 3) HANDLING OUTLIERS                        6)  FEATURE SELECTION WITH          9) FEATURE ENGINEERING
# 4) CATEGORICAL ENCODING                         DECISION TREE                   10) EXPLORATORY DATA ANALYSIS
# 5) NORMALIZATION AND STANDARDIZATION                                            11) PROGRAMMING LANGUAGE --> PYTHON,R
```
