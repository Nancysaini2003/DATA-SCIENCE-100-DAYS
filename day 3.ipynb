```python
# PROBALILITY : probabality measures the likelihood of a particular outcome or event occurring.
# it is typically expressed as a no. between 0 and 1 , where 0 indicates impossible (event will not occur))
# and 1 indicates certainty(event will occur).

# FORMULA : P(A) = TOTAL OF TIMES A OCCURS/TOTAL NO. OF POSSIBLE OUTCOMES
```

```python
# RANDOM VARIABLES : A random varible x is a function that assigns a real no.
# to each outcome in the sample space of a random experiment.

# there are a two types of random variable :
# 1) DISCRETE RANDOM VARIABLE : A random variable that takes on a countable no. of distnict values.
# EXAMPLE : [1,2,3,4,5,6]
# 2) CONTINUOUS RANDOM VARIABLE : A random variable that can take on any value within a given range or interval.
# EXAMPLE : [1------>10 RANGE]
```

```python
# PROBABILITY DISTRIBUTION : probability distributions describe how the probabilities of different outcomes are 
# distributed over the sample space of a random variable.
# 1) DISCRETE PROBABILITY DISTRIBUTIONS
# 2) CONTINUOUS PROBABILITY DISTRIBUTIONS 
```

```python
# PROBABILITY DISTRIBUTION FUNCTIONS : A probability distribution is the mathematical function that gives the probabilityes
# of the occurrence of different possibles outcomes for an experiment.

# there are three types of the probability distribution functions.
# 1) probability density functuin(PDF)
# 2) probability MASS functuin(PMF) 
# 3) cumulative density functuin(CDF)
```

```python
# NORMAL DISTRIBUTION : It known as the gussian distribution, is a probability that is symmetric about 
# the mean,showing that data near the mean are more frequent in occurrence than data far from the mean.
# In graph form ,normal distribution will appear as a bell curve.

# FORMULA : F(x) = 1//sigma root 2pie * e-1/2(x-µ)2/sigma.
```

```python
# STANDARD NORMAL DISTRIBUTION : 
# 1) the standard normal distribution,also known as the z distribution or z score, is a special case of the 
# the normal distribution.
# 2) mean (µ) of 0 and a standard deviation (sigma) of 1.
```

```python
#covariance and correlation

# 1) COVARIANCE : signifies the direction of the linear relationship between the two variables.by direction we 
# mean if the variables are directly proportional or inversely proportional to each other.
# 2) (increasing the value of one variable of one variable might have a positive or negative impact on the value of the other variable).

# FORMULA : COV(X,Y)=sum(xi-mean of x)*(yi-mean of y)//

# THREE TYPE OF COVARIANCE :
# 1) POSITIVE 
# 2) NEGATIVE 
# 3) ZERO 
 
```

```python
# CORRELATION : analysis is a method of statistical evaluation used to the strength of a relationship between two,numerically measured,continuous variable.
# FORMULA : correlation = cov(x,y)/sigma(x)*sigma(y)

# 0 to 0.5 ----> positive weak
# 0.5 to 1 ----> positive strong
# 0 to -0.5 -----> negative weak
# -0.5 to -1 -----> strong positive

```

```python
# HYPOTHESIS TESTING : it is a part of statistical analysis where we test the assumptions made regarding a population parameter.
# it is generally used when we were to compare a single group with an external standard and two or more groups with each other.

# TERMINOLOGY USED IN HYPOTHESIS TESTING :
# 1) NULL HYPOTHESIS : null hypothesis is a statistical theory that suggests there is no statistical significance exists between the populations. it is denoted by HO and reads as H-naught.
# EXAMPLE : A=B <=,>=
# 2) ALTERNATIVE HYPOTHESIS : an alternative hypothesis suggests there is a significant difference between the population parameters . it could be greater or smaller. 
#    basically,it is the contrast of the null hypothesis. it is denoted by Ha or H1.
# EXAMPLE : A<B,>,!=

# THERE ARE THREE TYPES OF HYPOTHESIS TESTING : 
# 1) Z-TEST (AVERAGE)
# 2) T-TEST (AVERAGE)
# 3) CHI-SQUARE TEST :
#                      1) CHECK THE GOODNESS
#                     2) RELATION FIND OUT 
    
# STEPS OF HYPOTHESIS TESTING :
# 1) state null(HO) and alternative(H1) hypothesis : HO must always contain equality(=).H1 always contains difference(!=,>,<).
# 2) choose level of significance(a) : 
# 3) find critical values
# 4) find test statistic
# 5) draw your conclusion
```
